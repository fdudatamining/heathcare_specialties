```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import *
```

```python
df = pd.read_csv('specialty_correlations.csv', header=None, skiprows=1, converters={'2': float})
```

```python
df
```

```python
def combination_to_matrix(df, a=0, b=1, c=2):
    d = df.pivot(a,b,c)
    d = d.add(d.T, fill_value=0).fillna(0)
    return d
d = combination_to_matrix(df)
```

```python
plt.pcolor(d)
plt.show()
```
